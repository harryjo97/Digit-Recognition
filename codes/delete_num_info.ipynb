```python
def avg_digit(input_imgs,th):
    
    img = np.zeros((10,784))

    for i in range(10):
        digit_data = input_imgs[input_imgs['digit']==i]
        digit = digit_data.iloc[0,3:]
        for j in range(1,digit_data.shape[0]-1):
            digit += digit_data.iloc[j,3:].values 
        digit = np.divide(digit,digit_data.shape[0])
        digit[digit< th]=0
        img[i] = digit

    return img
```

```python
a = avg_digit(train_copy,0)

fig, ax = plt.subplots(2,5,sharex=True,sharey=True,figsize=(20,8))
n=0

for row in range(2):
    for col in range(5):
        ax[row,col].imshow(  a[n].reshape(28,28),cmap='gray' )
        ax[row,col].axis('off')
        n += 1
```

```python
# find the overlapping part of input images with label and the image of label
def overlap(input_imgs,label):
    inputs = input_imgs.copy()
    output_imgs = inputs
    for i in range(output_imgs.shape[0]):

        arr_label = label[inputs.iloc[i,1]].copy()
        arr_label[arr_label>output_imgs.iloc[i,3:]] = 0
        output_imgs.iloc[i,3:] -= arr_label
        """for j in range(output_imgs.iloc[:,3:].shape[1]):
            if (output_imgs.iloc[i,j+3]>=label[inputs.iloc[i,1]][j]) : 
                output_imgs.iloc[i,j+3] -= label[inputs.iloc[i,1]][j]"""
        if i%100==0 : print(i, end=' ')

    return output_imgs
```

```python
clean_letter = overlap(train_copy,avg_digit(train_copy,th))
```

```python
fig, ax = plt.subplots(5,3,sharex=True,sharey=True,figsize=(20,20))

digit = avg_digit(train_copy,th)

for row in range(5):
    n = train_copy.iloc[row,1]

    raw = train_copy.iloc[row,3:].values.astype(np.int)
    ax[row,0].imshow(raw.reshape(28,28),cmap='gray')
    ax[row,0].axis('off')
    ax[row,0].set_title('letter : {}  digit : {}'.format(train_copy.iloc[row,2],n),color='w',fontsize=20)

    digit_img = digit[n] 
    ax[row,1].imshow(digit_img.reshape(28,28),cmap='gray')
    ax[row,1].axis('off')
    ax[row,1].set_title('digit : {}'.format(n),color='w',fontsize=20)

    tmp = clean_letter.iloc[row,3:].values.astype(np.int)
    ax[row,2].imshow(tmp.reshape(28,28),cmap='gray')
    ax[row,2].axis('off')
    ax[row,2].set_title('{} + {} - {}'.format(clean_letter.iloc[row,2],n,clean_letter.iloc[row,1]),color='w',fontsize=20)
```
